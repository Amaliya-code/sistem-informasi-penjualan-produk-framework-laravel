```python
from google_play_scraper import reviews, Sort
import csv

result, _ = reviews(
    'com.dafturn.mypertamina',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=200,
    filter_score_with=None
)

filename = 'ulasan_mypertamina.csv'

with open(filename, 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(
        f,
        fieldnames=['userName', 'score', 'at', 'content']
    )

    writer.writeheader()

    for review in result:
        writer.writerow({
            'userName': review['userName'],
            'score': review['score'],
            'at': review['at'],
            'content': review['content']
        })

print(f"Berhasil menyimpan {len(result)} ulasan ke '{filename}'")
```

```python
import pandas as pd
from transformers import pipeline

# Membaca file CSV hasil scraping
df = pd.read_csv('ulasan_mypertamina.csv')

# Menampilkan 5 data pertama
df.head()
```

```python
classifier = pipeline(
    "sentiment-analysis",
    model="w11wo/indonesian-roberta-base-sentiment-classifier"
)
```

```python
df['sentiment'] = df['content'].astype(str).apply(
    lambda x: classifier(x[:512])[0]['label']
)

df.head()
```

```python
df.to_csv('hasil_sentiment_mypertamina.csv', index=False)
```

```python
df['sentiment'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['sentiment'].value_counts().plot(kind='bar')

plt.title('Analisis Sentimen MyPertamina')
plt.xlabel('Sentiment')
plt.ylabel('Jumlah')

plt.show()
```
